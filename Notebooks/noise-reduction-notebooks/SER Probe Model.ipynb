```python
import os
import pandas as pd

# üîß Step 1: Set the path to your CREMA-D wav files
crema_dir = '/content/drive/MyDrive/FYP/ser-selective-enhancement/data/clean'
output_csv = '/content/drive/MyDrive/FYP/ser-selective-enhancement/data/labels.csv'

# üé≠ Step 2: Mapping CREMA-D emotion codes to full names
emotion_map = {
    'ANG': 'angry',
    'DIS': 'disgust',
    'FEA': 'fearful',
    'HAP': 'happy',
    'NEU': 'neutral',
    'SAD': 'sad'
}

# üß† Step 3: Approximate A/V/D values per emotion
avd_values = {
    'angry':     [0.85, -0.70, 0.75],
    'disgust':   [0.60, -0.60, 0.35],
    'fearful':   [0.90, -0.80, 0.30],
    'happy':     [0.80,  0.90, 0.65],
    'neutral':   [0.50,  0.00, 0.50],
    'sad':       [0.25, -0.65, 0.40]
}

# üìÑ Step 4: List .wav files
files = [f for f in os.listdir(crema_dir) if f.endswith('.wav')]

# üîç Step 5: Extract emotion from filename and assign A/V/D
rows = []
for f in files:
    try:
        code = f.split('_')[2].upper()  # e.g., 'SAD'
        emotion = emotion_map.get(code, 'unknown')
        a, v, d = avd_values.get(emotion, [0.0, 0.0, 0.0])
    except:
        emotion = 'unknown'
        a, v, d = 0.0, 0.0, 0.0
    rows.append([f, emotion, a, v, d])

# üíæ Step 6: Save to labels.csv
df = pd.DataFrame(rows, columns=['filename', 'emotion', 'arousal', 'valence', 'dominance'])
df.to_csv(output_csv, index=False)

print(f"‚úÖ labels.csv with {len(df)} entries saved to:\n{output_csv}")

# Optional: Preview
df.head()
```

```python
import os
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
def concordance_correlation_coefficient(y_true, y_pred):
    x_mean = torch.mean(y_true)
    y_mean = torch.mean(y_pred)
    cov = torch.mean((y_true - x_mean) * (y_pred - y_mean))
    x_var = torch.var(y_true)
    y_var = torch.var(y_pred)
    return (2 * cov) / (x_var + y_var + (x_mean - y_mean)**2 + 1e-8)

class CCCLoss(nn.Module):
    def forward(self, pred, target):
        ccc = concordance_correlation_coefficient(target, pred)
        return 1 - ccc  # we want to maximize CCC
```

```python
class SERProbeModel(nn.Module):
    def __init__(self):
        super(SERProbeModel, self).__init__()
        self.net = nn.Sequential(
            nn.Conv1d(1, 16, kernel_size=5, padding=2),
            nn.ReLU(),
            nn.MaxPool1d(2),
            nn.Conv1d(16, 32, kernel_size=5, padding=2),
            nn.ReLU(),
            nn.MaxPool1d(2),
        )
        self.fc = nn.Sequential(
            nn.Linear(32 * 25, 128),  # adjust to match input length
            nn.ReLU(),
            nn.Linear(128, 3)  # Arousal, Valence, Dominance
        )

    def forward(self, x):
        x = self.net(x)
        x = x.view(x.size(0), -1)
        return self.fc(x)
```

```python
def load_data(feature_dir, labels_df, lld_index):
    X = []
    Y = []
    missing_files = 0
    insufficient_llds = 0

    for file in os.listdir(feature_dir):
        if not file.endswith(".csv"):
            continue
        fpath = os.path.join(feature_dir, file)
        try:
            df = pd.read_csv(fpath, sep=';', comment='@')
        except:
            continue

        if df.shape[1] <= lld_index:
            insufficient_llds += 1
            continue

        series = df.iloc[:, lld_index].values[:100]  # truncate or pad
        if len(series) < 100:
            series = np.pad(series, (0, 100 - len(series)))
        else:
            series = series[:100]
        X.append(series)

        label_row = labels_df[labels_df['filename'] == file.replace('.csv', '.wav')]
        if label_row.empty:
            missing_files += 1
            continue
        Y.append(label_row[['arousal', 'valence', 'dominance']].values[0])

    if len(X) == 0 or len(Y) == 0:
        return None, None

    X = np.array(X)
    Y = np.array(Y)
    return X, Y
```

```python
def train_probe_model(X, Y, epochs=25):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model = SERProbeModel().to(device)
    criterion = CCCLoss()
    optimizer = optim.Adam(model.parameters(), lr=1e-4)

    X_train, X_val, Y_train, Y_val = train_test_split(X, Y, test_size=0.2)
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train.reshape(-1, 1)).reshape(X_train.shape)
    X_val = scaler.transform(X_val.reshape(-1, 1)).reshape(X_val.shape)

    for epoch in range(epochs):
        model.train()
        inputs = torch.tensor(X_train, dtype=torch.float32).unsqueeze(1).to(device)
        targets = torch.tensor(Y_train, dtype=torch.float32).to(device)
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, targets)
        loss.backward()
        optimizer.step()

    model.eval()
    with torch.no_grad():
        val_inputs = torch.tensor(X_val, dtype=torch.float32).unsqueeze(1).to(device)
        val_targets = torch.tensor(Y_val, dtype=torch.float32).to(device)
        preds = model(val_inputs)
        cccs = [concordance_correlation_coefficient(preds[:, i], val_targets[:, i]).item() for i in range(3)]
    return cccs
```

```python
labels = pd.read_csv('/content/drive/MyDrive/FYP/ser-selective-enhancement/data/labels.csv')
results = []

for lld_idx in range(2):
    print(f"\nTraining with LLD {lld_idx}")
    X, Y = load_data('/content/drive/MyDrive/FYP/ser-selective-enhancement/features/clean_lld', labels, lld_idx)
    if X is None or len(X) == 0:
        print(f"Skipping LLD {lld_idx} (no valid data)")
        continue
    try:
        cccs = train_probe_model(X, Y)
        results.append({
            'LLD_Index': lld_idx,
            'Arousal': cccs[0],
            'Valence': cccs[1],
            'Dominance': cccs[2]
        })
    except Exception as e:
        print(f"Error training LLD {lld_idx}: {e}")

df = pd.DataFrame(results)
df.to_csv('/content/drive/MyDrive/FYP/ser-selective-enhancement/outputs/probe_ccc_results_clean.csv', index=False)
```
